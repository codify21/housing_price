### 3(a) Missing Values 
Finding out which variables have mssing values

```python
var_with_na = [ var for var in data.columns if (data[var].isnull().sum())>1]

for var in var_with_na:
    print(var , np.round(data[var].isnull().mean(),3)*100,'%missing values')
```

#### Relationship b/w missing values and sales price

```python
def analyse_na_value(df,var):
    df = df.copy()
    
    #Making variable that indicates 1 if observation was missing or 0 otherwise
    df[var] = np.where(df[var].isnull(),1,0) 
    
    #calculate the median salesprice where the information is missing or present
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.show()
    
for var in var_with_na:
    analyse_na_value(data,var)
```

### 3(b) Numerical Variables

```python
num_vars = [var for var in data.columns if data[var].dtypes != 'O']
len(num_vars)
```

```python
data[num_vars].head()
```

### 3(b)(1) Temporal variables

```python
year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]
year_vars
```

```python
for var in year_vars:
    print(var , data[var].unique())
    print()
```

```python
data.groupby('YearRemodAdd')['SalePrice'].median().plot()
plt.ylabel('Median House Price')
plt.title('Change in Houseprice with years')
```

```python
#To analyse b/w house price & year variables in more detail
def analyse_year_vars(df,var):
    df= df.copy()
    
    df[var]= df['YrSold']-df[var]
    
    plt.scatter(df[var],df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in year_vars:
    if var != 'YrSold':
        analyse_year_vars(data,var)
```

   ### 3(b)(2) : Discrete Variables : shows a finite no. of values

```python
discrete_vars = [var for var in num_vars if len(data[var].unique())<20 and var not in year_vars+['Id'] ]
print(len(discrete_vars))
discrete_vars
```

```python
data[discrete_vars].head()
```

#### Relationship b/w discrete variables and sales price

```python
def analyse_discrete(df,var):
    df= df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data,var)
```

   ### 3(b)(3) : Continous Variables 

```python
#every other remaining variable except for year & discrete will be considered as continous for now

count_vars = [var for var in num_vars if var not in discrete_vars+year_vars+['Id']]
print(count_vars)
print(len(count_vars))
```

```python
data[count_vars].head()
```

#### Relationship b/w continous variables and sales price

```python
def analyse_continous(df,var):
    df= df.copy()
    df[var].hist(bins=20)
    plt.ylabel('No. of houses')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in count_vars:
    analyse_continous(data,var)
```

### Applying transformations to convert the skewed distribution into more like gaussion distribution

```python
def analyse_transformed_continous(df,var):
    df =df.copy()
    
    #log dont take 0 or negative values , so transformation applied to only positive & non-zero variables
    if 0 in df[var].unique():
        pass
    else:
        df[var]=np.log(df[var])
        df[var].hist(bins=20)
        plt.ylabel('Number of Houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
for var in count_vars:
    analyse_transformed_continous(data,var)
```

```python
#Exploring relationship b/w house price and transformed variables
def transform_analyse_continous(df,var):
    df =df.copy()
    
    if 0 in df[var].unique():
        pass
    else:
        df[var]=np.log(df[var])
        df['SalePrice']= np.log(df['SalePrice'])
        plt.scatter(df[var],df['SalePrice'])
        plt.ylabel('Saleprice')
        plt.xlabel(var)
        plt.show()
        
for var in count_vars:
    if var != 'SalePrice':
        transform_analyse_continous(data,var)
```

### 3(d) Outliers

```python
#Making boxplots to visualise outliers
def find_outliers(df,var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column = var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
        
for var in count_vars:
    find_outliers(data,var)
        
```
